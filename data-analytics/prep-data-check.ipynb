```python
import matplotlib.pyplot as plt

def column_filter(c):
    if 'diff' in c:
        return True
    else:
        return False
filtered_columns = list(filter(column_filter, df.columns.tolist()))

plt.matshow(df.set_index('timestamp')[filtered_columns].corr())
```
